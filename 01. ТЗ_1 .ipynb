Вам предстоит поработать со списком возрастов посетителей фитнес-клубов age_lst

**Задача 1.** Рассчитайте средний возраст по всему списку age_lst

**Задача 2.** Рассчитайте средний возраст для посетителей, которые моложе 50 лет

**Задача 3.** Оберните предыдущий код в функцию, которая на вход принимает список и пороговое значение

**Задача 4.** Рассчитайте среднее значение для всех элементов списка, за исключением тех, которые лежат вне интервала [Q1-1,5*IQR; Q3+1,5*IQR]

**Задача 5 (со звёздочкой).** Оберните предыдущий код в функцию, которая принимает на вход список и рассчитывает средний возраст, исключая из расчётов выбросы

```python
age_lst = [23, 45, 18, 80, 22, 25, 34, 37, 41, 88, 19, 20, 18, 45, 41, 39, 31, 32, 22, 28, 37]
```

```python
# Задача 1. Найдем средний возраст всех посетителей из списка age_lst
avg_age = sum(age_lst)/len(age_lst)
print(f'Средний возраст посетитей {round(avg_age, 0)}')
```

```python
# Задача 2. Вычислим средний аозраст для посетителей младше 50 лет 
sum_age=0
len_age=0
for i in age_lst:
    if i < 50:
        sum_age+=i
        len_age+=1
avg_age = sum_age/len_age
print(f'Средний возраст посетителей младше 50 лет {round(avg_age, 0)}')
```

```python
# Задача 3.Создаем функцию рассчета среднего возраста с порогом
def calc_avg_age(lst, thresh):
    sum_age=0
    len_age=0
    for i in lst:
        try:
            if i < thresh:
                sum_age+=i
                len_age+=1
        except TypeError:
            print(f'{i} Это не числовое значение')
    try:
        avg_age = sum_age/len_age
        print(f'Средний возраст посетителей младше {thresh} лет {round(avg_age, 0)}')
        return round(avg_age, 0)
    except:
        print(f'Список должен содержать быть хотя бы одно значение меньше {thresh}')
```

```python
calc_avg_age(age_lst, 50)
```

```python
# Задача 4.
import numpy as np
```

```python
# Рассчитаем медиану, первый и третий квартили
q1 = np.percentile(age_lst, 25)
q2 = np.percentile(age_lst, 50)
q3 = np.percentile(age_lst, 75)
print(q1, q2, q3)
```

```python
# Определим межквартильный размах, верхнюю и нижнюю границу
iqr = q3 - q1
lower_bound = q1-1.5*iqr
upper_bound = q3+1.5*iqr
print(iqr)
print(lower_bound)
print(upper_bound)
```

```python
# Рассчитаем средний возраст, исключив выбросы за пределами нижней и верхней границ
sum_age=0
len_age=0
for i in age_lst:
    if (i<upper_bound) and (i>lower_bound):
        sum_age+=i
        len_age+=1
avg_age = sum_age/len_age
print(f'Новое значение среднего возраста посетителей равно {round(avg_age, 0)}')
```

```python
# Задача 5.
# Создадим функцию определения среднего возраста без выбросов
def calc_avg_age(lst):
    try:
        q1 = np.percentile(lst, 25)
        q3 = np.percentile(lst, 75)
        iqr = q3 - q1
        lower_bound = q1-1.5*iqr
        upper_bound = q3+1.5*iqr
        sum_age=0
        len_age=0
        for i in lst:
           if (i<upper_bound) and (i>lower_bound):
               sum_age+=i
               len_age+=1
        avg_age = sum_age/len_age
        return round(avg_age, 0)
    except:
        print(f'Список не должен содержать нечисловые значения или быть пустым')
```
